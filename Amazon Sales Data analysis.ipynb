### 1. Product Analysis: Analyze the distribution of product categories, sizes, and quantities sold to identify popular products. 

```python
Qty = data1.groupby('Category')['Qty'].sum().tolist()
category = data1.Category.unique().tolist()
df = pd.DataFrame({"Category" : category, "Qty" : Qty})
df.sort_values(by = 'Qty', ascending=False, inplace = True)
ax = sns.barplot(x = df['Category'], y = df['Qty'])
for i in ax.containers:
    ax.bar_label(i)
```

 - In this bar chart, you can see the overall sales of each category of product.
 - The most bought product is 'Socks', while the least bought product is 'Watch'.

### 2. Fulfillment Analysis: Investigate the fulfillment methods used and their effectiveness in delivering orders

```python
data1.groupby('Category')['Fulfilment'].value_counts()
```

- As you can see in the bar chart, here is the distribution of fulfilment methods used for each category of products

```python
plt.figure(figsize = (13, 6))
ax = sns.countplot(data = data1, x = 'Category', hue = 'Fulfilment', palette = ['c', 'm'])
for i in ax.containers:
    ax.bar_label(i)
```

```python
count = data1.Fulfilment.value_counts().tolist()
```

```python
efficiency_amazon = count[0] / np.sum(count) * 100
print(f"efficiency of amazon : {efficiency_amazon}")
print(f"efficiency of merchant : {100 - efficiency_amazon}")
```

```python
plt.pie(
    x = [efficiency_amazon, 100 - efficiency_amazon],
    labels = ['Amazon', 'Merchant'],
    autopct = "%1.2f%%",
    explode = [0, 0.1],
    colors = ['orange', 'pink']
)
plt.show()
```

- As you can see, Fulfilment by Amazon is the most efficienct fulfilment method

### 3. Sales Overview: Understand the overall sales performance, trends, and patterns over time

```python
data1.groupby('Category').Amount.sum()
```

```python
fig, ax = plt.subplots(9, 1, figsize = (15, 50))
# sns.scatterplot(x = data1.index,y = 'Amount', hue = 'Category', data = data1)
cat_list = data1.Category.unique().tolist()
cat_list = iter(cat_list)
for i in range(9):
    a = next(cat_list)
    df = data1[data1['Category'] == a].copy()
    axs = sns.lineplot(x = df.index,y = 'Amount', data = df, ax = ax[i])
    axs.set_title(f'Sales of {a}')
    axs.grid()
```

### 6. Business Insights: Provide actionable insights and recommendations based on the analysis to optimize sales strategies, improve customer satisfaction, and enhance overall business performance.

```python
state = data1['ship-state'].unique().tolist()[1]
df5 = pd.DataFrame(data1[data1['ship-state'] == state].groupby('Size')['Category'].value_counts())
df5
```

 - As you can see here, In the state of Maharashtra,  the most bought product is the Shirt of size 'XL'. 
 - Similarly we can find the popular product in all the states.
 - Hence, to expand the business, the popular product should be manufacture more and should be sold only in that state where the sales are product are highest

```python
plt.figure(figsize = (15, 15))
states = data1['ship-state'].unique().tolist()
x1 = []
y = []
for state in states:
    df5 = pd.DataFrame(data1[data1['ship-state'] == state].groupby('Size')['Category'].value_counts())
    x = pd.DataFrame(df5.iloc[np.argmax(df5)])
    y.append(df5.iloc[np.argmax(df5)][0])
    x1.append(f'{state} : {x.columns.values[0][0]}, {x.columns.values[0][1]}')
    
ax = sns.barplot(y = x1, x = y)
for i in ax.containers:
    ax.bar_label(i)
```

- Hence Here is the list of products that are popular in each states which can be manufactured and sold more often in that state to generate more revenue.
- Hence in this way business can be increased in this way
